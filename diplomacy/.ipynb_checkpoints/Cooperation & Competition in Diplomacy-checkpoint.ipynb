One helpful way to think about the connections in the game and inform analysis is to create a graph with all the connections between countries and oceans. 

```python
# Create graph w/ countries & edges to use for distance calculations

G = nx.Graph()
map_connections = Map().dest_with_coasts

for source_location in map_connections.keys():
    for dest_location in map_connections[source_location]:
        G.add_edge(source_location, dest_location)
```

```python
unit_locations = game.get_units() 
```

We can use this graph, along with data on the positions of each army/fleet, to compute a "threat matrix" where we calculate the distance from each power's supply centers to other armies/fleets. 

We do this by taking the [Djikstra path](https://en.wikipedia.org/wiki/Dijkstra%27s_algorithm) along the graph and take the minimum distance for each power / supply center combination. 

```python
def compute_threat_matrix(current_power):
  
    threat_matrix = []

    for supply_center in game.get_centers()[current_power]:
        for power in unit_locations.keys():
              if power != current_power:
                    for unit in unit_locations[power]:
                        enemy_unit_location = unit.split(' ')[1]
                        threat_matrix.append([supply_center, 
                                              power, 
                                              nx.dijkstra_path_length(G, supply_center, enemy_unit_location)])

    threat_df = pd.DataFrame(threat_matrix, columns = ['supply_center', 'enemy_power', 'distance'])
    threat_matrix = pd.pivot_table(data=threat_df.groupby(['supply_center', 'enemy_power']).agg('distance').min().reset_index(), index='supply_center', columns='enemy_power', values='distance', aggfunc=np.sum)

    return threat_matrix
```

Let's start by analyzing Austria - in particular, let's compute the threat matrix and compare this to the map. 

```python
austria_threat_matrix = compute_threat_matrix('AUSTRIA')
print(austria_threat_matrix)
```

```python
# Add distances for each power and rank in ascending order to see top threats
austria_threat_matrix.sum().sort_values()
```

Looking at the map, unsurprisingly, the top threat to Austria is Italy, with Germany and Russia not far behind. England is the furthest away and not necessarily something that has to be dealt with immediately.

In short: Several supply centers in the Balkans are available early on, but you need to figure out alliances with Italy, Germany, and/or Russia. 

It's helpful at this stage to understand the options available to a given player. Here's a function utilizing the game API to parse all available moves. 

```python
def get_power_possible_orders(current_power):
  
    power_possible_orders = []
  
    possible_orders = game.get_all_possible_orders()
    controlled_locations = game.get_orderable_locations(current_power)
  
    for i in controlled_locations:
        power_possible_orders.append(possible_orders[i])

    return power_possible_orders
```

## Section 2: Analyzing strategies

As you play more, a natural question arises around how to decide who to cooperate vs. compete with.

To help answer these questions, let's take a look at some game data and see if there are patterns that we can pull out. Access to the dataset mentioned in the DipNet paper was contingent on a confidentiality agreement, so in the interest of making something broadly available for this class, we found an alternative dataset publicly available on data.world. This is a smaller dataset, but nonetheless provides a good starting point. We spent some time translating this game history to the diplomacy package format and were able to parse the results of several thousand games into the dataset available here.

While the DipNet paper relies on graph embeddings to train an LSTM, we take a simpler approach for this assignment. Instead of focusing on specific moves, we focus more generally on the payoffs for cooperating vs. competing with a given player for a given time. 

- We define payoffs as the net increase or decrease in total supply centers in the turn following the current turn, and these payoffs are conditional on the set of players that a given player allies with or attacks in that turn. 
- We define alliance by evaluating the possible moves for a given player in relationship to other players:
    - If a player can attack another player and does, we call that a conflict
    - If a player can attack another player and does not, we call that an alliance

This is done to facilitate simplicity in analysis. Note that this does not factor in situations where 
- Players who are not adjacent to each other create an alliance (this is potentially a very flawed assumption)
- Players can attack each other but do not (e.g. because they do not have strength and assume attack would fail, another potentially flawed assumption) 
- Players have to get in position to attack another player 
- Players introduce a red herring conflict meant to throw off other players
- Etc. 

Note that it also does not consider long term or other emergent dynamics of player behavior. If for instance, a player is initially very aggressive, they might capture a number of supply centers in the next turn, but other players may decide to ally against them and take them out.

It also does not factor in that changes in supply centers may not be causally connected - e.g. a player may take supply centers that have nothing to do with a particular relationship. 

That all being said, let's take an initial look at some data and see what we are able to pull out. 

```python
df = pd.read_csv('all_game_payoffs.csv')
```

```python
df
```

Our data here captures the payoff (net supply center changes in next turn) based on a given power, turn/phase, and the set of players they cooperate with vs. attack.  Note that `set()` denotes an empty set, meaning no cooperation or attacks at least by our definition. 

Note that we have a few instances of very long games where by some fluke the game lasted a century - this is uncommon. 

We'll start by computing the distribution of payoffs for a given player to cooperate or attack another player for a given turn. For example, we will return the distribution of payoff for Italy for cooperating vs. fighting with Austria in Fall 1901 movement. 

Let's write a function to return the relevant distributions of cooperative vs. attack payoffs for different players of the payoff dataframe. For now, we will ignore the different combinations of player alliances, and just focus on if there was ANY cooperation or fighting. Our function has 4 arguments: 
1. `data` - the input dataframe, which will always be the `df` dataframe containing payoffs. 
2. `power1` - the active power. If we are assessing whether Italy should cooperate or fight with Austria, `power1` would be `'ITALY'`. 
3. `power2` - the opposing power. In our example, this would be `'AUSTRIA'`.
4. `turn` - the current phase. 

Make the function return 2 numpy arrays containing the payoffs for power 1 cooperating with vs. attacking power 2, for a given turn/phase.

```python
def compute_payoff_distributions(data, power1, power2, turn):

    ### BEGIN SOLUTION

    cooperate_array = data[(data['power'] == power1) & 
                    (data['coop'].str.contains(power2)) & 
                    (data['phase'] == turn)]['payoff'].to_numpy()

    attack_array = data[(data['power'] == power1) & 
                    (data['attack'].str.contains(power2)) & 
                    (data['phase'] == turn)]['payoff'].to_numpy()


    ### END SOLUTION
    return cooperate_array, attack_array
```

Here's a test to make sure you are generating these arrays properly. 

```python
italy_coop_austria_distribution, italy_attack_austria_distribution = compute_payoff_distributions(df,
                                                                                                  'ITALY', 
                                                                                                  'AUSTRIA', 
                                                                                                  'F1901M')

assert len(italy_coop_austria_distribution) == 2463
assert len(italy_attack_austria_distribution) == 1048
assert round(np.mean(italy_coop_austria_distribution), 2) == 0.71
assert round(np.mean(italy_attack_austria_distribution), 2) == 0.99

### BEGIN HIDDEN TESTS

eng_coop_france_dist, eng_attack_france_dist = compute_payoff_distributions(df,
                                                                            'ENGLAND',
                                                                            'FRANCE',
                                                                            'F1905M'
                                                                           )

assert len(eng_coop_france_dist) == 1672
assert len(eng_attack_france_dist) == 1121
assert round(np.mean(eng_coop_france_dist), 2) == 0.01
assert round(np.mean(eng_attack_france_dist), 2) == 0.08

### END HIDDEN TESTS
```

```python
plt.figure(figsize=(15,10))
sns.histplot(italy_coop_austria_distribution, label= 'Italy cooperates with Austria', alpha=0.5, discrete=True)
sns.histplot(italy_attack_austria_distribution, label = 'Italy attacks Austria', alpha=0.5, discrete=True)
plt.grid()
plt.legend()
plt.show()
```

A few interesting things stick out to us reviewing this visualization. Namely, we can see that it much more common for Italy to cooperate with Austria, but there does appear to be a higher payoff (at least in the short run) if Italy decides to attack Austria. That does not, however, factor in the consequences of playing aggressively - remember, if a player is aggressive at the beginning of the game, other players might ally to try and take them down. 

Let's compare this to the payoffs for Austria associated with cooperating vs. attacking Italy...

```python
austria_coop_italy_distribution, austria_attack_italy_distribution = compute_payoff_distributions(df,
                                                                                                  'AUSTRIA', 
                                                                                                  'ITALY', 
                                                                                                  'F1901M')

```

```python
plt.figure(figsize=(15,10))
sns.histplot(austria_coop_italy_distribution, label= 'Austria cooperates with Italy', alpha=0.5, discrete=True)
sns.histplot(austria_attack_italy_distribution, label = 'Austria attacks Italy', alpha=0.5, discrete=True)
plt.grid()
plt.legend()
plt.show()
```

```python
round(np.mean(austria_coop_italy_distribution))
```

```python
round(np.mean(austria_attack_italy_distribution))
```

Interestingly, it seems like there is a more equal distribution of payoffs for Austria in considering whether to cooperate with or attack Italy, though it is much more common for them to cooperate than to fight, at least during this particular turn.
